```python
# Q.9

IT_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27}
age = [22, 19, 24, 25, 26, 24, 25, 24]

# Length of the set IT_companies
print("Length of IT_companies:", len(IT_companies))

# Add Twitter
IT_companies.add('Twitter')
print("Added Twitter:", IT_companies)

# Insert multiple IT companies at once
IT_companies.update(['Netflix', 'LinkedIn'])
print("Added multiple companies:", IT_companies)

# Remove one of the companies
IT_companies.remove('Oracle')
print("After removing Oracle:", IT_companies)

"""Remove and Discard both remove the specified element.
If the element does not exist, remove would raise an error,
but discard would not."""

print()
print("A union B: ", A.union(B))
print("A intersection B: ", A.intersection(B))
print("Is A a subset of B: ", A.issubset(B))
print("Are A & B disjoint sets: ", A.isdisjoint(B))
print("Join A with B: ", A.union(B))
print("Join B with A: ", B.union(A))
print("Symmetric difference between A and B: ", A.symmetric_difference(B))
print()

del A
del B

age_set = set(age)
print("Age list: ", age)
print("Age set: ", age_set)
print("Length of age list: ", len(age))
print("Length of age set: ", len(age_set))
```

```python
class Employee:
  numEmps = 0
  sumSalary = 0
  def __init__(self, name, family, salary, department):
    self.name = name
    self.family = family
    self.salary = salary
    self.department = department
    Employee.numEmps += 1
    Employee.sumSalary += salary

  def avgSalary(self):
    return Employee.sumSalary / Employee.numEmps

class FulltimeEmployee(Employee):
  def __init__(self, name, family, salary, department):
    super().__init__(name, family, salary, department)

emp1 = Employee("Alice", "Smith", 60000, "HR")
emp2 = FulltimeEmployee("Bob", "Johnson", 80000, "Engineering")
emp3 = FulltimeEmployee("Charlie", "Brown", 90000, "Marketing")

print(f"Employee 1 (Alice) - Name: {emp1.name}, Family: {emp1.family}, Salary: {emp1.salary}, Department: {emp1.department}")
print(f"Employee 2 (Bob) - Name: {emp2.name}, Family: {emp2.family}, Salary: {emp2.salary}, Department: {emp2.department}")
print(f"Employee 3 (Charlie) - Name: {emp3.name}, Family: {emp3.family}, Salary: {emp3.salary}, Department: {emp3.department}")

print(f"Total number of employees: {Employee.numEmps}")
print(f"Average salary of all employees: {emp1.avgSalary()}")
```
